```python
#this is copy from python2 code ,not well in python 3
import urllib.request as ulr
fhand=ulr.urlopen('http://www.py4inf.com/code/romeo.txt')

counts=dict()
for line in fhand:
    words=line.spilt()
    for word in words:
        counts[word]=counts.get(word,0)+1

    
print(counts)
```
